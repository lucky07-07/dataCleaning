```python
Q1=df1["avg_glucose_level"].quantile(0.25)
print("Q1:",Q1)
Q2=df1["avg_glucose_level"].quantile(0.5)
print("Q2:",Q2)

Q3=df1["avg_glucose_level"].quantile(0.75)
print("Q3:",Q3)

Q4=df1["avg_glucose_level"].quantile(1)
print("Q4:",Q4)

#Calculate IQR
iqr=Q3-Q1
print("iqr:",iqr)

#Lower Limit
ll=Q1-(iqr*1.5)
print("ll:",ll)
#Upper Limit
ul=Q3+(iqr*1.5)
print("ul:",ul)


```

```python
df1[(df1["bmi"]<ll)| (df1["bmi"]>ul)]
```

```python
df1 = df1.copy()
df1["bmi"] = df1["bmi"].clip(lower=22.855, upper=169.475)
df1
```

```python
sns.boxplot(x=df1["bmi"])
```

```python
Q1 = df1["bmi"].quantile(0.25)
Q3 = df1["bmi"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows where "bmi" is outside the range
df1 = df1[(df1["bmi"] >= lower_bound) & (df1["bmi"] <= upper_bound)]
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df1["bmi"])
plt.show()
```
